# Filtering

#### * The filter() function extracts elements from an iterable list, tuple etc. When the condition met function returns True or else False.

```python
dataset.head(9)
```

```python
dataset.loc[(dataset.Marks>=90) & (dataset.Admission==2019)]
```

```python
# I want to filter those student records who belongs to lucknow
# True value meets our filter criteria and false doesn't
dataset['SAddress']== 'Lucknow'
```

```python

filt=(dataset['SAddress']== 'Lucknow')
```

```python
filt
```

```python
dataset[filt]
```

```python
Topper=(dataset['Marks']>=90)| (dataset['Specialization'] == 'Finance')
```

```python
dataset.loc[Topper,['Id','StudentName','Specialization','Courses','Marks']]
```
